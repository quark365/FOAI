```python
import random
import time
from ipywidgets import GridBox, Button, Layout
from IPython.display import display
from aima3.logic import expr


X, O, BLANK = 'X', 'O', ' '


Situation, Action, Square, Player, Mark = expr('Situation'), expr('Action'), expr('Square'), expr('Player'), expr('Mark')


winning_combinations = [

    [(Square(1, 1), Mark(X)), (Square(1, 2), Mark(X)), (Square(1, 3), Mark(X))],
    [(Square(2, 1), Mark(X)), (Square(2, 2), Mark(X)), (Square(2, 3), Mark(X))],
    [(Square(3, 1), Mark(X)), (Square(3, 2), Mark(X)), (Square(3, 3), Mark(X))],

    [(Square(1, 1), Mark(X)), (Square(2, 1), Mark(X)), (Square(3, 1), Mark(X))],
    [(Square(1, 2), Mark(X)), (Square(2, 2), Mark(X)), (Square(3, 2), Mark(X))],
    [(Square(1, 3), Mark(X)), (Square(2, 3), Mark(X)), (Square(3, 3), Mark(X))],

    [(Square(1, 1), Mark(X)), (Square(2, 2), Mark(X)), (Square(3, 3), Mark(X))],
    [(Square(1, 3), Mark(X)), (Square(2, 2), Mark(X)), (Square(3, 1), Mark(X))]
]


def is_winning(situation, player):
    for combination in winning_combinations:
        if all((expr_in_situation(situation, square, mark) and mark == Mark(player)) for square, mark in combination):
            return True
    return False


def expr_in_situation(situation, square, mark):
    return expr('In')(square, situation) & expr('Mark')(square, mark)


def play_tic_tac_toe():
    board = [[BLANK, BLANK, BLANK] for _ in range(3)]
    current_player = X
    game_over = False


    def on_button_click(button, row, col):
        nonlocal current_player, game_over
        if not game_over and button.description == BLANK:
            button.description = current_player
            board[row][col] = current_player
            if check_win(board, current_player):
                with output_widget:
                    print(f"Player {current_player} wins!")
                game_over = True
                disable_buttons()
            elif all(mark != BLANK for row in board for mark in row):
                with output_widget:
                    print("It's a draw!")
                game_over = True
                disable_buttons()
            else:
                current_player = O if current_player == X else X
                if current_player == O:
                    time.sleep(1)
                    computer_move()
        else:
            with output_widget:
                print("That square is already taken. Try again.")
                time.sleep(2)
                output_widget.clear_output()

    buttons = [Button(description=BLANK, layout=Layout(width='50px', height='50px')) for _ in range(9)]


    for i, button in enumerate(buttons):
        row = i // 3
        col = i % 3
        button.on_click(lambda _, r=row, c=col: on_button_click(buttons[r * 3 + c], r, c))


    grid_box = GridBox(buttons, layout=Layout(grid_template_columns='auto auto auto', width='200px'))
    display(grid_box)
    output_widget = Output()
    display(output_widget)


    def disable_buttons():
        for button in buttons:
            button.disabled = True


    def computer_move():
        nonlocal current_player, game_over
        _, move = minimax(board, O)
        r, c = move
        buttons[r * 3 + c].description = current_player
        board[r][c] = current_player
        if check_win(board, current_player):
            with output_widget:
                print(f"Player {current_player} wins!")
            game_over = True
            disable_buttons()
        elif all(mark != BLANK for row in board for mark in row):
            with output_widget:
                print("It's a draw!")
            game_over = True
            disable_buttons()
        else:
            current_player = X if current_player == O else O


    def minimax(board, player):
        if check_win(board, X):
            return -1, None
        elif check_win(board, O):
            return 1, None
        elif all(mark != BLANK for row in board for mark in row):
            return 0, None

        best_score = float('-inf') if player == O else float('inf')
        best_move = None

        for i in range(3):
            for j in range(3):
                if board[i][j] == BLANK:
                    board[i][j] = player
                    score, _ = minimax(board, X if player == O else O)
                    board[i][j] = BLANK

                    if player == O:
                        if score > best_score:
                            best_score = score
                            best_move = (i, j)
                    else:
                        if score < best_score:
                            best_score = score
                            best_move = (i, j)

        return best_score, best_move

def check_win(board, player):
    for row in board:
        if all(mark == player for mark in row):
            return True
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True
    if all(board[i][i] == player for i in range(3)):
        return True
    if all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

play_tic_tac_toe()
```
